```python
import pandas as pd
df=pd.read_excel('Combined_Data.xlsx')
#converting to all caps
df['COURSE_CODE'] = df['COURSE_CODE'].str.upper()
df['MEETING_DAYS'] = df['MEETING_DAYS'].str.upper()
df['CLASS_TYPE']=df['CLASS_TYPE'].str.upper()
df['INSTRUCTOR']=df['INSTRUCTOR'].str.upper()
df['COURSE_NAME']=df['COURSE_NAME'].str.upper()
df['ROOM_NUMBER']=df['ROOM_NUMBER'].str.upper()
df = df.rename(columns={'SEATS_AVAILABALE': 'SEATS_AVAILABLE'})
df.head()
```

```python
#drop unwanted columns
df = df.drop(columns=['CLASS_CODE_NUMBER', 'ONLINE_TELIVISED','GENERAL_EDUCATION','CLASS_NOTES','ROOM_NUMBER'])
#Handling filled number that ends with /
df['FILLED'] = df['FILLED'].str.rstrip('/')
df.head()
```

```python
#Total rows
row_count = len(df)
print(row_count)
```

```python
#Drop nulls in SEAT_FILLED and FILLED
df = df.dropna(subset=['SEATS_FILLED'])
df = df.dropna(subset=['FILLED'])
row_count = len(df)
print(row_count)
```

```python
#Checking null in filled
isNull=df[df['FILLED'].isnull()]
isNull.count()
```

```python
#Counting filled number that end with /
filtered_df = df[df['FILLED'].str.endswith('/')]
print(filtered_df)
```

```python
#Checking total if it starts with /
filtered_df = df[df['TOTAL'].str.startswith('/')]
filtered_df.count()
```

```python
#Handling total that starts with /
df['TOTAL'] = df['TOTAL'].str.lstrip('/')
# Handling data of total that starts witj / 
filtered_df = df[df['TOTAL'].str.startswith('/')]
filtered_df.count()
```

```python
#Deleting values having inconsistent entries
filtered_df = df[df['FILLED'].str.contains('-')]
filtered_df
#Row 2583
df = df.drop(index=2583)
df= df.replace('II', 0)
df= df.replace('1-3', 3)
df= df.replace('1-6', 3)
df = df.drop(index=5996)
```

```python
df['TOTAL'].unique()
```

```python
#Converting Filled and Total to int type
df['FILLED'] = df['FILLED'].astype(int)
df['TOTAL'] = df['TOTAL'].astype(int)
df['SEATS_AVAILABLE'] = df['TOTAL'] - df['FILLED']
df['SEATS_AVAILABLE']=df['SEATS_AVAILABLE'].astype(int)
#Calculating SEATS_AVAILABLE

df.head(10)
```

# Data Cleaning

```python
#Handling SECTION_NUMBER
print(df['SECTION_NUMBER'].isnull().sum())
df['SECTION_NUMBER'].fillna(0, inplace=True)
print(df['SECTION_NUMBER'].isnull().sum())
```

```python
df['UNITS'].fillna(3, inplace=True)
print(df['INSTRUCTOR'].isnull().sum())
```

```python
#Adding New column
df['YEAR']=pd.Series(dtype=int)
#Assigning Year to the data
df['YEAR'] = '20' + df['DATE_RANGE'].str[-2:]
unique_count = df['YEAR'].value_counts()
print(unique_count)
```

```python
#Adding New column
df['YEAR']=pd.Series(dtype=int)
#Assigning Year to the data
df['YEAR'] = '20' + df['DATE_RANGE'].str[-2:]
unique_count = df['YEAR'].value_counts()
print(unique_count)
```

```python
#Adding new column SEMESTER
df['SEMESTER']=pd.Series(dtype=int)
def get_semester(date_range):
    if date_range.startswith('01'):
        return 'SPRING'
    elif date_range.startswith('08'):
        return 'FALL'
    else:
        return None
#Setting SEMESTER value
df['SEMESTER'] = df['DATE_RANGE'].apply(get_semester)
```

```python
df['SEMESTER'].unique()
```

```python
#Removing repetitive coulumns of seats
df = df.drop(columns=["SEATS_FILLED","FILLED","TOTAL"])
```

```python
#Finding Unique values in UNITS column
value_counts = df['UNITS'].value_counts()
# print the result
print(value_counts)
```

```python
#Removing repetitive coulumns of seats
df = df.drop(columns=["DATE_RANGE"])
```
