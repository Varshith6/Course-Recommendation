```python
import pandas as pd
df=pd.read_excel('Combined_Data.xlsx')
#converting to all caps
df['COURSE_CODE'] = df['COURSE_CODE'].str.upper()
df['MEETING_DAYS'] = df['MEETING_DAYS'].str.upper()
df['CLASS_TYPE']=df['CLASS_TYPE'].str.upper()
df['INSTRUCTOR']=df['INSTRUCTOR'].str.upper()
df['COURSE_NAME']=df['COURSE_NAME'].str.upper()
df['ROOM_NUMBER']=df['ROOM_NUMBER'].str.upper()
df = df.rename(columns={'SEATS_AVAILABALE': 'SEATS_AVAILABLE'})
df.head()
```

```python
#drop unwanted columns
df = df.drop(columns=['CLASS_CODE_NUMBER', 'ONLINE_TELIVISED','GENERAL_EDUCATION','CLASS_NOTES','ROOM_NUMBER'])
#Handling filled number that ends with /
df['FILLED'] = df['FILLED'].str.rstrip('/')
df.head()
```

```python
#Total rows
row_count = len(df)
print(row_count)
```

```python
#Drop nulls in SEAT_FILLED and FILLED
df = df.dropna(subset=['SEATS_FILLED'])
df = df.dropna(subset=['FILLED'])
row_count = len(df)
print(row_count)
```

```python
#Checking null in filled
isNull=df[df['FILLED'].isnull()]
isNull.count()
```

```python
#Counting filled number that end with /
filtered_df = df[df['FILLED'].str.endswith('/')]
print(filtered_df)
```
